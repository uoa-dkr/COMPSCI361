```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

## lines

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
fig = plt.figure()        # For all Matplotlib plots, we start by creating a figure and an axes
ax = plt.axes()
```

```python
fig,ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'red', linestyle = 'dashdot', label = 'sin(x)')   # modify color and linetype 
plt.xlim(-5,5)
plt.ylim(-.5,.5)          # or axis([-5,5,-.5,.5]) or axis('tight')/axis('equal')
plt.title('a sin graph')
plt.xlabel('x-axis')
plt.ylabel('y-aixs')
plt.legend()              # label can be applied here
plt.show()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
xlabel='x', ylabel='sin(x)',
title='A Simple Plot')
```

## scatter plot

```python
plt.plot(x,np.sin(x),'o')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label= f"marker='{marker}'")
plt.legend(numpoints = 1)
plt.show()
```

```python
plt.plot(x,np.sin(x),'-ok')      # '-' is line, 'o' is point, 'k' is black
plt.show()
```

```python
plt.scatter(x,np.sin(x))
plt.show()
```

```python
r1 = rng.rand(100)
r2 = rng.rand(100)
plt.scatter(r1, r2, s = r1*1000, c = r2, alpha = .5)   # s is size and c is color
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,s=200*features[3], c=iris.target, cmap='viridis') 
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar()
plt.show()
```
